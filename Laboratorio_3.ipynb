4) obtenga la media aritmética de la diferencia (a-b).

```python
#Crea el array
a = np.array([[1,4,2,7]])
b = np.array([[1,3,2,9]])
#se obtiene la diferencia
diferencia = (a-b)
# se obtiene la media
media = np.mean([diferencia])
print("La media es", media)
```

 ## 1. ¿Cuántos Municipios tienen más de 100000
habitantes?.

```python
# Filtra los municipios con poblacion > 100 000
municipios_mas_100 = dataset[dataset['Poblacion'] > 100000]
# Hace un conteo de los municipios
cantidad_municipios = municipios_mas_100.shape[0]
print("La cantidad de municipios con mas de 100 000 son",cantidad_municipios)
```

 ## 2.  Realice una gráfica de barras sobre la polbación de cada
ciudad, ordenela de menor a mayo y responda:

```python
import matplotlib.pyplot as plt
#Ordenamos el df
df_ordenado = dataset.sort_values(by='Poblacion')
# Hacemos la grafica
plt.figure(figsize=(100, 100))
plt.barh(df_ordenado['Municipio'], df_ordenado['Poblacion'], color='skyblue')
plt.xlabel('Población')
plt.ylabel('Municipio')
plt.title('Población de cada Municipio')
plt.grid(axis='x')
plt.show()
```

### 2.1 Ciudad más poblada

```python
# Busca cual es la segunda ciudad mas poblada
municipio = df_ordenado['Municipio'].iloc[-2]
print("a. ¿Cuál es la segunda ciudad más poblada?:", municipio)
```

### 2.2 Municipo de granada con 5000 habitantes

```python
municipio_granada = df_ordenado[df_ordenado['Municipio'] == 'Granada'].index.item() + 1
print("b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?:", municipio_granada)
```

## 3. ¿Cuántos municipios de Extremadura tienen más de 5000 **
habitantes?.

```python
# Filtra Extremadura
municipios_mas_5000 = dataset[(dataset['Poblacion'] > 5000) & (dataset['Provincia'] == 'Extremadura')]
cantidad_municipios_extremadura = municipios_mas_5000.shape[0]
print("¿Cuántos municipios de Extremadura tienen más de 5000 habitantes?",cantidad_municipios_extremadura )
```

 ## 4.  ¿Cuál es el municipio situado más al Norte?

```python
# Buscamos a el municipio que tiene mayor coordenada en Y
municipio_norte = dataset.sort_values(by='Y', ascending=False)
municipio_norte = municipio_norte.loc[municipio_norte['Y'].idxmax()]
nombre_municipio = municipio_norte['Municipio']
nombre_provincia = municipio_norte['Provincia']
poblacion = municipio_norte['Poblacion']
print("El muunicipio mas al norte es: ",nombre_municipio,"y se encuentra en la provincia: ",nombre_provincia,"y tiene una poblacion de: ",poblacion)
```

## 7) Seleccione al azar cincuenta municipios diferentes de entre los diponiblesen el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
data_ordenada = dataset.sort_values(by='Provincia')
data_ordenada['row_number'] = data_ordenada.groupby('Provincia').cumcount() + 1
primeros_50_datos = data_ordenada[data_ordenada['row_number'] <= 50]
primeros_50_datos.head()

```
