```python
def get_variables_scale_type(dataset):
  columnas = dataset.columns
  categoricas = []
  continuas = []
  discretas = []
  for col in columnas :
  col_type = dataset[col].dtype()
  n = len(dataset[col].unique())
    if(col_type == 'object'):
      categoricas.append(col)
    elif((col_type == 'int64') or (col_type == 'float64')):
      n= len(dataset[col].unique())
      if(n>30):
        continuas.append(col)
      else
      discretas.append(col)
return continuas, discretas categoricas
```

```python
continuas, discretas, categoricas = get_variables_scale_type(dataset)
```
